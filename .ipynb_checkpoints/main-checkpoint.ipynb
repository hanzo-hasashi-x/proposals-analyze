```python
def parse_project_bids(project):
    base = [project['id'], project['title'], project['description'], project['budget']['minimum'],
        project['budget']['maximum'], project['submitdate']]
    
    project_bids = []
        
    for bid in project['bids']:
        row = [*base]
        
        reputation = bid['reputation']['entire_history']['category_ratings']
        
        row += [bid['score'], bid['highlighted'], bid['sealed'], bid['description'], bid['submitdate'], 
                bid['amount'], bid['period'], reputation['quality'], reputation['communication'], 
                reputation['professionalism'], bid['award_status'] == 'awarded']
        
        project_bids.append(row)
    
    return project_bids
```

```python
data = []
for project in projects:
    
    project_bids = parse_project_bids(project)
        
    if [bid for bid in project_bids if bid[-1]]: # checking if there is awarded bid, if not skip
        data+=project_bids
```

### Experiment with features

```python
def num_transform(df):
    df['project/bid_amount_proportion'] = df['bid_amount'] / df['project_min_amount']
    df['bid_delay'] = (df['bid_time_submitted'] - df['project_time_submitted']) / (3600*24*7) 
    
    df = df.drop(['bid_amount', 'project_min_amount', 'project_max_amount', 
                  'project_time_submitted', 'bid_time_submitted'], axis=1)
    return df
```

```python
data = num_transform(data)
```

#### Transform text data in features

```python
def get_proportion(df, column_name):
    proportions = []

    projects_titles_words = df[column_name].apply(lambda x: x.split()).apply(set)
    bid_descrs_words = df['bid_description'].apply(lambda x: x.split()).apply(set)

    for project_title_set, bid_descr_set in zip(projects_titles_words, bid_descrs_words):
        words_checked = [word in bid_descr_set for word in project_title_set]
        proportions.append(sum(words_checked) / len(words_checked))

    return proportions

greetings = ['good afternoon', 'good morning', 'how are you', 'how do you do', 'dear sir',
             'good night', 'hi', 'hello', 'pleased to meet you']

def transform_text_columns(df):
    
    df['bid_description'] = df['bid_description'].fillna('')

    df['same_words_title/bid_proportion'] = get_proportion(df, 'project_title')
    df['same_words_descr/bid_proportion'] = get_proportion(df, 'project_description')

    df['bid_descr_len'] = df['bid_description'].apply(lambda x: len(x.split()))
    
    proj_descr_len = df['project_description'].apply(lambda x: len(x.split()))
    df['proj/bid_lens_proportion'] = df['bid_descr_len'] / proj_descr_len

    df['bid_greetings'] = df['bid_description'].apply(lambda x: x.lower()).apply(
        lambda x: any([x.find(greeting) != -1  for greeting in greetings]))
    
    df = df.drop(['project_title', 'project_description', 'bid_description'], axis=1)
    
    return df
```

```python
data = transform_text_columns(data)
```

#### Splitting data in datasets (train, test)

```python
data_projects = data.groupby('project_id').groups
```

```python
from sklearn.utils import shuffle

train_size = .8

group_indexes = list(data_projects.values())
group_indexes = shuffle(group_indexes, random_state=42)

to_index = int(train_size * len(data_projects))
train_indexes = [index for project in group_indexes[:to_index]
                  for index in project]
test_indexes = [index for project in group_indexes[to_index:]
                  for index in project]
```

```python
X = data.drop(['project_id', 'bid_award_status'], axis=1)
project_ids = data['project_id'].values

X_num, X_cat = X.select_dtypes(include=['number']).values, X.select_dtypes(exclude=['number']).values

y = data['bid_award_status'].to_numpy()

X_train_num, X_test_num  = X_num[train_indexes], X_num[test_indexes]
X_train_cat, X_test_cat = X_cat[train_indexes], X_cat[test_indexes]

y_train, y_test = y[train_indexes], y[test_indexes]

train_project_ids, test_project_ids = project_ids[train_indexes], project_ids[test_indexes]
```

#### Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()

X_train_num = scaler.fit_transform(X_train_num)
X_test_num = scaler.transform(X_test_num)
```

```python
X_train, X_test = np.c_[X_train_num, X_train_cat], np.c_[X_test_num, X_test_cat]
```

### Training algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC, SVC

from functools import partial

from sklearn.metrics import accuracy_score, precision_score, recall_score

def get_metrics(y,pred):
    for metric in [accuracy_score, precision_score, recall_score]:
        if metric.__name__ == 'accuracy_score':
            result = metric(y, pred)
        else:
            result = metric(y, pred, zero_division=0)
        print(metric.__name__[:-6], f'{round(result*100,2)}%')

trained_models = [] # for future hard voting

models = [partial(LogisticRegression, max_iter=10_000), RandomForestClassifier, GradientBoostingClassifier,
          partial(SVC, probability=True), KNeighborsClassifier, GaussianNB]
for model in models:
        
    try:
        model = model(class_weight='balanced', random_state=42)
    except:
        if model != KNeighborsClassifier and model != GaussianNB:
            model = model(random_state=42)
        else:
            model = model()
    
    print(type(model).__name__)
        
    model.fit(X_train, y_train)
    trained_models.append(model)
    
    pred = model.predict(X_test)
    
    get_metrics(y_test,pred)
        
    print()
```

- for precision: the best is Gradient Boosting. With this model, only for 4 of 100 project will be given result, but one of two of such predictions will be correct

- for recall: the best is Logistic Regression. In the predictions of such model, it will be predicted 90% of awarded bids, but only 13.19% of all positive predictions will be correct

#### Looking at the accuracy for each project

```python
def get_accuracies(y,pred,project_ids):
    top_accuracy, top_2_accuracy = [],[]
    for project_id in set(project_ids):
        indexes = project_ids == project_id
        
        if sum(indexes) < 10: continue
        
        y_project,pred_project = y[indexes], pred[indexes]
        
        # choosing the best bid (along predictions)
        top_accuracy.append( y_project[np.argmax(pred_project)] )
        
        # choosing the top 2 best bids (along predictions)
        sorted_pred = np.argsort(pred_project)[:-2 -1:-1]
        top_2_accuracy.append( any(y_project[ sorted_pred ]) )
        
    top_accuracy = np.mean(top_accuracy) * 100
    top_2_accuracy = np.mean(top_2_accuracy) * 100

    return top_accuracy, top_2_accuracy
```

```python
for model in trained_models:
    pred = model.predict_proba(X_test)[:, 1]
    
    top_accuracy, top_2_accuracy = get_accuracies(y_test, pred, test_project_ids)

    print('model name: {}, top accuracy: {:.2f}%, top 2 accuracy: {:.2f}%'.format(
        type(model).__name__,top_accuracy,top_2_accuracy))
    
print('Projects test number', len(set(test_project_ids)))
```

SVC gives the most accurated prediction, if to take first, and from 2 first it's necessary to take Gradient Boosting

### Deeper training of model (SVC - Support Vector Classification)

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform

svc = SVC(class_weight='balanced', random_state=42, probability=True)

params_distributions = {"gamma": reciprocal(.0001, 0.5), "C": uniform(1,15)}
rnd_search_cv = RandomizedSearchCV(svc, params_distributions, n_iter=10, verbose=2, cv=3, scoring='f1')
rnd_search_cv.fit(X_train, y_train)
```

```python
round(rnd_search_cv.best_score_*100,2)
```

```python
best_estimator = rnd_search_cv.best_estimator_
best_estimator
```

```python
pred = best_estimator.predict(X_test)
get_metrics(y_test,pred)
```

### Hard voting

```python
from sklearn.ensemble import VotingClassifier

vot_clf = VotingClassifier([(type(model).__name__, model) for model in trained_models],
                           voting='hard')
vot_clf.fit(X_train, y_train)

pred = vot_clf.predict(X_test)
get_metrics(y_test, pred)
```

### Soft voting

```python
from sklearn.ensemble import VotingClassifier

vot_clf = VotingClassifier([(type(model).__name__, model) for model in trained_models],
                           voting='soft')
vot_clf.fit(X_train, y_train)

pred = vot_clf.predict(X_test)
get_metrics(y_test, pred)
```

it appears that combining of all networks resulted to the best precision, but the prediction will given only for 6 projects of 100 and 3 of them will correct

### Features importance analyze

```python
best_estimator = grid_search_cv.best_estimator_
best_estimator
```

```python
pred = best_estimator.predict(X_test)
get_metrics(y_test,pred)
```

```python
rf = best_estimator

forest_importances = pd.Series(rf.feature_importances_, index=X.columns)

forest_importances.sort_values().plot.bar()
```

Currently it's possible to match some main properties to look at
- Bid communication and quality (previous reviews)
- Bid period (how time will it take in the opinion of freelancer)
- Bid delay (how much of time is passed)
- Same words in bid from title
- The proportion between project min price and bid proposed one

Bid score isn't remembered because possibly it's only the sum of other properties made by freelancer.com, and which defines the position of bid

```python
for column in ['bid_score', 'bid_period', 'bid_delay', 'same_words_title/bid_proportion', 
               'project/bid_amount_proportion']:
    categ_col,bins = pd.qcut(data[column], 5, duplicates='drop',
                       labels=False, retbins=True)
    
    classes = sorted(set(categ_col))
    classes_n = [sum( data['bid_award_status'][categ_col == class_] ) for class_ in classes]

    bins_str = [f'{round(bins[i],4)}-{round(bin_,4)}' for i,bin_ in enumerate(bins[1:])]    
    plt.figure(figsize=(10, 4))
    plt.barh(bins_str, classes_n)
    
    plt.title(column)
    plt.show()
```

from these bars, it's possible to conclude that
- The best price for work is in interval 0-2 times more than min
- Than it's not necessary to repeat all title words, it will be sufficient to use 10%
- Than bigger period than less possibility to be awarded
- it's sufficient to be in 20 first (bid score)
- There is anomaly with very little bid delay, when bids receive 2 times less than in others bins while the bids are written before others, which can be explained by bots which don't write a lot of sens information which could interest the owner of project

### Pipeline

```python
from sklearn.compose import ColumnTransformer

cols_num, cols_cat = X.select_dtypes(include=['number']).columns, X.select_dtypes(exclude=['number']).columns

full_pipeline = ColumnTransformer([
    ('scaler', scaler, cols_num),
    ('cat_pipeline', k_bins, cols_cat)
])
full_pipeline = full_pipeline.fit(X)
```

```python
from sklearn.pipeline import Pipeline
import joblib

full_pipeline_with_estimator = Pipeline([
    ('full_pipeline', full_pipeline),
    ('estimator', vot_clf)
])

joblib.dump(full_pipeline_with_estimator, 'estimators/ml_bids_estimator.json')
```
